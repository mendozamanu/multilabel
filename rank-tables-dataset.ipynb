```python
import pandas as pd
import numpy as np
import math
```

```python
br = pd.read_csv('csv/medidas-iterativa-4-2-br.csv', sep=';', decimal=',')
print br
```

```python
cc = pd.read_csv('csv/medidas-iterativa-4-2-cc.csv', sep=';', decimal=',')
print cc
```

```python
lp = pd.read_csv('csv/medidas-iterativa-4-2-lp.csv', sep=';', decimal=',')
print lp
```

```python
ml = pd.read_csv('csv/medidas-iterativa-4-2-mlknn.csv', sep=';', decimal=',')
print ml
```

```python
tables = [
    br,
    cc,
    lp,
    ml
]
```

```python
tmp=np.zeros(len(tables))
tmp2=np.zeros(len(tables))
tmp3=np.zeros(len(tables))
tmp4=np.zeros(len(tables))
tmp5=np.zeros(len(tables))
tmp6=np.zeros(len(tables))
tmp7=np.zeros(len(tables))


tm=np.zeros(len(br['Dataset']))
for tb in tables:
    tb['Acc Rank'] = tm
    tb['HL Rank'] = tm
    tb['Avg prec Rank'] = tm
    tb['ROC Rank'] = tm
    tb['F1 micro Rank'] = tm
    tb['Recall Rank'] = tm
    tb['F1 macro Rank'] = tm
j=0
p=1
for i in range(0, len(br['Dataset'])):
    j=0
    for tb in tables:
        #print tb['Accuracy'][i]
        #tenemos q hacer ranks de todos los tb['Accuracy'][i]
        #print tb
        tmp[j]=tb['Accuracy'][i]
        tmp2[j]=tb['Hamming Loss'][i]
        tmp3[j]=tb['Avg precision'][i]
        tmp4[j]=tb['ROC AUC'][i]
        tmp5[j]=tb['f1 score (micro)'][i]
        tmp6[j]=tb['Recall'][i]
        tmp7[j]=tb['f1 score (macro)'][i]
        j+=1
        
    #print tmp7
    p=1
    while True:
        j=0
        idx=np.argmax(tmp)
        idx2=np.argmin(tmp2)
        idx3=np.argmax(tmp3)
        idx4=np.argmax(tmp4)
        idx5=np.argmax(tmp5)
        idx6=np.argmax(tmp6)
        idx7=np.argmax(tmp7)
        
        if math.isnan(tmp[j]):
            for rw in range(0, len(tables)):
                tables[rw]['Acc Rank'][i] = float('nan')
        else:
            tmp[idx] = 0
            tables[idx]['Acc Rank'][i] = p
        #----------------------------------------------------
        if math.isnan(tmp2[j]):
            for rw in range(0, len(tables)):
                tables[rw]['HL Rank'][i] = float('nan')
        else:
            tmp2[idx2] = np.inf
            tables[idx2]['HL Rank'][i] = p
        #----------------------------------------------------
        if math.isnan(tmp3[j]):
            for rw in range(0, len(tables)):
                tables[rw]['Avg prec Rank'][i] = float('nan')
        else:
            tmp3[idx3] = 0
            tables[idx3]['Avg prec Rank'][i] = p
        #----------------------------------------------------
        if math.isnan(tmp4[j]):
            for rw in range(0, len(tables)):
                tables[rw]['ROC Rank'][i] = float('nan')
        else:
            tmp4[idx4] = 0
            tables[idx4]['ROC Rank'][i] = p
        #----------------------------------------------------
        if math.isnan(tmp5[j]):
            for rw in range(0, len(tables)):
                tables[rw]['F1 micro Rank'][i] = float('nan')
        else:
            tmp5[idx5] = 0
            tables[idx5]['F1 micro Rank'][i] = p
        #----------------------------------------------------
        if math.isnan(tmp6[j]):
            for rw in range(0, len(tables)):
                tables[rw]['Recall Rank'][i] = float('nan')
        else:
            tmp6[idx6] = 0
            tables[idx6]['Recall Rank'][i] = p
        #----------------------------------------------------
        #print idx7
        if math.isnan(tmp7[j]):
            for rw in range(0, len(tables)):
                tables[rw]['F1 macro Rank'][i] = float('nan')
        else:
            tmp7[idx7] = 0
            tables[idx7]['F1 macro Rank'][i] = p
        #----------------------------------------------------
       
        p+=1
        j+=1
        if p>4:
            break
```

```python
br['Mean'] = br[['Acc Rank', 'HL Rank', 'Avg prec Rank', 'ROC Rank', 'F1 micro Rank', 'Recall Rank', 'F1 macro Rank']].mean(axis=1)
```

```python
cc['Mean'] = cc[['Acc Rank', 'HL Rank', 'Avg prec Rank', 'ROC Rank', 'F1 micro Rank', 'Recall Rank', 'F1 macro Rank']].mean(axis=1)
```

```python
lp['Mean'] = lp[['Acc Rank', 'HL Rank', 'Avg prec Rank', 'ROC Rank', 'F1 micro Rank', 'Recall Rank', 'F1 macro Rank']].mean(axis=1)
```

```python
ml['Mean'] = ml[['Acc Rank', 'HL Rank', 'Avg prec Rank', 'ROC Rank', 'F1 micro Rank', 'Recall Rank', 'F1 macro Rank']].mean(axis=1)
```
